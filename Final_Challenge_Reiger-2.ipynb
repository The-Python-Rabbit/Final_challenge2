```python
print ("This is the Rock Paper Scissors program")
# The rules in Rock, Paper, Scissors are:

#    Rock beats scissors
#    Scissors beats paper
#    Paper beats rock
# If the user wins, then display "You win!"
# If they lose against the computer, then display "You lose." 
# If the computer and the user have the same play choice, display "It's a draw."


import random as randnum
user_score =0
computer_score = 0

while True:
    choice = (input("Enter (R)ock, (P)aper, (S)cissors (or 'q' to quit):"))
    choice1 = choice[0].lower()
    if choice1 == 'q':
        break
    if choice1 != 'r' and choice1 != 'p' and choice1 != 's':
        print ("please (R)ock, (P)aper, (S)cissors")
        continue
 
    
    comp_choice = randnum.randint(1,3)
#    print (comp_choice)
    if comp_choice == 1:
        print ("Computer chooses Rock")
    if comp_choice ==2:
        print ("Computer chooses Paper")
    if comp_choice == 3:
        print ("computer chooses Scissors")
        

    # 1 = rock, 2 = paper, 3= scissors
    if choice == 'r' and comp_choice == 3:
        print ("You win!")
        user_score +=1
    elif choice == 'p' and comp_choice == 1:
        print ("You win!")
        user_score +=1
    elif choice == "s" and comp_choice == 2:
        print ("You win!")
        user_score +=1
    elif choice == 's' and comp_choice == 1:
        print ("You lose.")
        computer_score +=1
    elif choice == 'p' and comp_choice == 3:
        print ("You lose")
        computer_score +=1
    elif choice == "r" and comp_choice == 2:
        print ("You lose")
        computer_score +=1
    else:
        print ("It's a draw")
        
print ("Program finished")
print ("You scored: ", user_score)
print ("Computer scored: ", computer_score)
    # print (choice1)
    # print (comp_choice)

```
